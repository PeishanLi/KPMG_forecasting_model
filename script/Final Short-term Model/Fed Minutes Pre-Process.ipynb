```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('fed_funds_predictions.csv')
```

```python
df.date[0].split('/')[2][:4] + '-' + df.date[0].split('/')[0] + '-' + '1'
```

```python
timestamp = []
for i in range(len(df)):
    timestamp.append(df.date[i].split('/')[2][:4] + '-' + df.date[i].split('/')[0] + '-' + '1')
timestamp
```

```python
df.date = timestamp
```

```python
df1 = pd.DataFrame()
timestamp1 = []
for i in range(1993,2023,1):
    for j in range(1,13,1):
        timestamp1.append(str(i)+'-'+str(j)+'-1')
```

```python
df1['date'] = timestamp1
df['date'] = pd.to_datetime(df.date)
df1['date'] = pd.to_datetime(df1.date)
```

```python
df = df.groupby('date').mean()
df.reset_index(inplace=True)
```

```python
df2 = df1.merge(df,how='left')
df2 = df2.fillna(method='ffill')
```

```python
df2.fed_funds_rate = df2.fed_funds_rate.shift(1)
```

```python
df2.to_csv('preprocessed_Fed_minutes.csv')
```
